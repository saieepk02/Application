```python
import nltk

from nltk.tokenize import word_tokenize

# Download required resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab') # Added to download the missing resource
nltk.download('averaged_perceptron_tagger_eng') # Added to download the missing resource

# Input sentence
text = "Natural Language Processing enables machines to understand human language."

# Tokenize the sentence
tokens = word_tokenize(text)

# HMM PoS Tagging using NLTK
pos_tags = nltk.pos_tag(tokens)

print("PoS Tagging using HMM (NLTK):")

for word, tag in pos_tags:
    print(word,"->",tag)
```

```python
import spacy

# Load English model (neural network-based)

nlp = spacy.load("en_core_web_sm") # Make sure it's installed

# Input sentence

text = "Natural Language Processing enables machines to understand human language."

# Process the text

doc = nlp(text)

print("\nPoS Tagging using Neural Model (SpaCy):")

for token in doc:
    print(f"{token.text} → {token.pos_} ({token.tag_})")
```
