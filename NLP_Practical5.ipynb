```python
#A. Syntactic Parsing using NLTK (CFG-based Parser)

print("Snehal Kedar")

import nltk

from nltk import CFG

# Define a simple grammar

grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N | Det Adj N | PN
VP -> V NP
Det -> 'a' | 'the'
N -> 'dog' | 'cat' | 'man'
Adj -> 'tall' | 'big'
V -> 'saw' | 'liked'
PN -> 'John' | 'Mary'
""")

# Create parser

parser = nltk.ChartParser(grammar)

# Input sentence

sentence = ['John', 'saw', 'a', 'tall', 'man']

# Parse and print trees

print("Syntactic Parse Trees using NLTK:")

for tree in parser.parse(sentence):
    print(tree)
    tree.pretty_print()
```

```python
print("Snehal Kedar")
print("Syntactic Parsing using SpaCy (Pre-trained Model)")

import spacy

# Load SpaCy's English model

nlp = spacy.load("en_core_web_sm")

# Input sentence

sentence = "John saw a tall man."

# Process the sentence

doc = nlp(sentence)

# Display syntactic dependencies

print("\nDependency Parse (SpaCy):")

for token in doc:
    print(f"{token.text:<10} {token.dep_:<12} {token.head.text:<10} POS: {token.pos_}")
```
