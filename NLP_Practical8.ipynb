```python
# Probabilistic Context Free Grammar (PCFG)

# Grammar with probabilities
grammar = {
    "S": [(["NP","VP"],1.0)],
    "NP": [(["Det","N"],0.6), (["N"],0.4)],
    "VP": [(["V","NP"],0.7), (["V"],0.3)],
    "Det": [(["the"],0.5), (["a"],0.5)],
    "N": [(["cat"],0.5), (["dog"],0.5)],
    "V": [(["chased"],0.6), (["slept"],0.4)]
}

# Sentence to parse
sentence = ["the","cat","chased","a","dog"]

# Example parse probability (manual rule sequence)
probability = 1

rules_used = [
    ("S",["NP","VP"],1.0),
    ("NP",["Det","N"],0.6),
    ("Det",["the"],0.5),
    ("N",["cat"],0.5),
    ("VP",["V","NP"],0.7),
    ("V",["chased"],0.6),
    ("NP",["Det","N"],0.6),
    ("Det",["a"],0.5),
    ("N",["dog"],0.5)
]

print("Rules Used:\n")

for lhs,rhs,p in rules_used:
    print(f"{lhs} -> {' '.join(rhs)}  [{p}]")
    probability *= p

print("\nSentence:", " ".join(sentence))
print("Total Probability:", probability)
```
