```python

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
print("Snehal Kedar")
# Download tokenizer models (only needed once)
nltk.download('punkt')
nltk.download('punkt_tab') # Added to resolve LookupError for punkt_tab

# Sample text
text = "Natural Language Processing is a fascinating field. It allows computers to understand human language."

# Sentence Segmentation
sentences = sent_tokenize(text)
print("\nSentence Segmentation:")
for idx, sentence in enumerate(sentences, 1):
    print(f"{idx}: {sentence}")

# Word Tokenization
print("\nWord Tokenization:")
for idx, sentence in enumerate(sentences, 1):
    words = word_tokenize(sentence)
    print(f"Sentence {idx} Tokens: {words}")
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
text = "NLP is interesting. It helps computers understand language."

# Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentence Segmentation:")
for i in range(len(sentences)):
    print(i+1, sentences[i])

# Word Tokenization
print("\nWord Tokenization:")
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    print("Sentence", i+1, "Token :", words)
```
