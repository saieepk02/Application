```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Download resources if not already done
nltk.download('punkt')
nltk.download('punkt_tab') # Added to download the missing resource

# Input text
text = "Machine learning is transforming the world with innovative applications in healthcare, finance, and education."

# Tokenize and convert to lowercase
tokens = word_tokenize(text.lower())

# Generate trigrams
tri = list(ngrams(tokens, 3))

# Display the trigrams
print("Generated Trigrams:")
for t in tri:
    print(t)

# Frequency count
freq = Counter(tri)

print("\nTrigram Frequencies:")
for t,c in freq.items():
    print(t,":",c)
```
